```python
import pandas as pd
```

```python
df= pd.read_excel("Data set (2).xlsx")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
numeric_columns = df.select_dtypes(include='number').columns
print("Numeric columns:", numeric_columns)
```

```python
for column in numeric_columns:
    df[column] = df[column].fillna(df[column].median())
```

```python
df.isnull().sum()
```

```python
data_mean= df[numeric_columns].mean()
print("Mean values are:\n",data_mean)
```

```python
data_median = df[numeric_columns].median()
print("Median values are:\n",data_median)
```

```python
data_mode = df[numeric_columns].mode()
print("Mode values are:\n",data_mode)
```

```python
# Calculating Range for each numeric column
range_value = df[numeric_columns].max() - df[numeric_columns].min()
print("Range is:\n", range_value)
```

```python
# Calculating Variance
variance_value = df[numeric_columns].var()
print("Variance is:\n", variance_value)
```

```python
# Calculating Standard Deviation
std_value = df[numeric_columns].std()
print("Standard Deviation is:\n", std_value)
```

```python
# Calculating skewness
skewness_value = df[numeric_columns].skew()
print("Skewness is:\n", skewness_value)
```

```python
# Calculating kurtosis
kurtosis_value = df[numeric_columns].kurt()
print("Kurtosis is:\n", kurtosis_value)
```
